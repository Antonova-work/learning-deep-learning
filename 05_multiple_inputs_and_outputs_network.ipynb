```python
# прогноз процента травмированных членов команды, победы и эмоционального состояния игроков (печаль/радость)
def neural_network(inputs, weights):
  pred = vect_mat_mul(inputs, weights)

  return pred
```

```python
# вычисление взвешенной суммы (попарно умножаем значение из inputs с весом из соответсвующей строки weights (зависит от итерации в vect_mat_mul))
def w_sum(a, b):
  assert(len(a) == len(b))

  output = 0

  for i in range(len(a)):
    output += (a[i] * b[i])

  return output
```

```python
# сохраняет соответсвующий вычисленный результат-прогноз из функции w_sum и проводит итерацию по вложенным спискам weights
def vect_mat_mul(vect, matrix):
  assert(len(vect) == len(matrix))

  output = [0, 0, 0]

  for i in range(len(vect)):
    output[i] = w_sum(vect, matrix[i])

  return output
```

```python
# веса (разный набор для разных прогнозов, но одинаковый состав (среднее число игр, сыгранных игроками; процент побед; число болельщиков))
weights = [[0.1, 0.1, -0.3], # прогноз процента травмированных игроков
           [0.1, 0.2, 0.0], # прогноз победы
           [0.0, 1.3, 0.1]] # прогноз эмоционального состояния игроков

# текущее среднее число игр, сыгранных, сыгранных игроками
toes = [8.5, 9.5, 9.9, 9.0]

# текущая доля игр, окончившихся победой
wlrec = [0.65, 0.8, 0.8, 0.9]

# число болельщиков (в миллионах)
nfans = [1.2, 1.3, 0.5, 1.0]

# запись первой игры в сезоне
inputs = [toes[0], wlrec[0], nfans[0]]

pred = neural_network(inputs, weights)
```

```python
# процент травмированных членов команды, победы, эмоционального состояния игроков (печаль/радость)
print(pred)
```
