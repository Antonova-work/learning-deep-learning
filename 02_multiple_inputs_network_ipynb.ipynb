## Код без использования NumPy

```python
# считаем скалярное произведение (умножаем попарно значения векторов и складываем их)
def w_sum(a, b):
  # утверждаем, что вектора равны
  assert(len(a) == len(b))
  output = 0
  for i in range(len(a)):
    output += (a[i] * b[i])
  return output
```

```python
# прогноз победы
def neural_network(inpt, weights):
  pred = w_sum(inpt, weights)
  return pred
```

```python
# веса
weights = [0.1, 0.2, 0]

# текущее число игр
games = [8.5, 9.5, 9.9, 9.0]
# текущая доля игр, окончившихся победой (процент)
wlrec = [0.65, 0.8, 0.8, 0.9]
# число болельщиков в миллионах
nfans = [1.2, 1.3, 0.5, 1.0]
```

```python
# первая игра в сезоне
inpt = [games[0], wlrec[0], nfans[0]]
pred = neural_network(inpt, weights)
print(pred)
```

### Замечание: наиболее существенным прогнозным признаком является wrecl, потому что он имеет наибольший вес (0.2), но наиболее существенный вклад в прогноз вносит games, потому что вход, объединяемый с весом, имеет самое большое значение.

## Код с NumPy

```python
import numpy as np
```

```python
# прогноз победы
def neural_network(inpt, weights):
  # "dot product" - скалярное произведение
  pred = inpt.dot(weights)
  return pred
```

```python
# веса
weights = np.array([0.1, 0.2, 0])

# текущее число игр
games = np.array([8.5, 9.5, 9.9, 9.0])
# текущая доля игр, окончившихся победой (процент)
wlrec = np.array([0.65, 0.8, 0.8, 0.9])
# число болельщиков в миллионах
nfans = np.array([1.2, 1.3, 0.5, 1.0])
```

```python
# первая игра в сезоне
inpt = np.array([games[0], wlrec[0], nfans[0]])
pred = neural_network(inpt, weights)
print(pred)
```
