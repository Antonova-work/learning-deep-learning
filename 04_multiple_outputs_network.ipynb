```python
# считаем вероятность победы/поражения путём умножения на разные веса
def ele_mul(number, vector):

  # number - вероятность победы/порожения
  # vector - вектор весов для нескольких выходов

  output = [0, 0, 0]

  assert(len(output) == len(vector))

  for i in range(len(vector)):
    output[i] = number * vector[i]

  # возвращает вектор с прогнозируемыми вероятностями
  return output
```

```python
# прогноз победы, процента травмированных членов команды и эмоционального состояния игроков (печаль/радость)
def neural_network(inputs, weights):
  pred = ele_mul(inputs, weights)

  return pred
```

```python
# вероятность победы/поражения
wlrec = [0.65, 0.8, 0.8, 0.9]

# веса
weights = [0.3, 0.2, 0.9]

# на вход только один параметр - вероятность победы/поражения
inputs = wlrec[0]

pred = neural_network(inputs, weights)
```
